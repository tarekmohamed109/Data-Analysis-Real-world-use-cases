# 3.. Analysing change in price of the stock overtime?

```python
# Sort the rows of the DataFrame in descending order based on the index,
# reset the index, and remove the 'index' column
data = bitcoin.sort_index(ascending=False).reset_index()
data.drop('index', axis=1, inplace=True)
```

```python
data
```

```python
# Create a subplot with 2 rows and 2 columns, and specify subplot titles
fig = make_subplots(rows=2, cols=2, subplot_titles=['Open', 'High', 'Low', 'Close'], shared_xaxes=True)

# Add a scatter plot for 'Open' prices in the first subplot
fig.add_trace(go.Scatter(x=bitcoin['Date'], y=bitcoin['Open'], name='Open'), row=1, col=1)

# Add a scatter plot for 'High' prices in the second subplot
fig.add_trace(go.Scatter(x=bitcoin['Date'], y=bitcoin['High'], name='High'), row=1, col=2)

# Add a scatter plot for 'Low' prices in the third subplot
fig.add_trace(go.Scatter(x=bitcoin['Date'], y=bitcoin['Low'], name='Low'), row=2, col=1)

# Add a scatter plot for 'Close' prices in the fourth subplot
fig.add_trace(go.Scatter(x=bitcoin['Date'], y=bitcoin['Close'], name='Close'), row=2, col=2)

# Customize the layout of the figure
fig.update_layout(height=800, width=990, title_text='Bitcoin Price')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Price')

# Display the figure
fig.show()
```

```python
# Create a Candlestick trace with the required data
trace = go.Candlestick(x=data['Date'], high=data['High'], open=data['Open'], close=data['Close'], low=data['Low'])

# Combine the trace into a list
candle_data = [trace]

# Define the layout of the figure
layout = {
    'title': 'Bitcoin Historical Price',
    'xaxis': {'title': 'Date'}
}

# Create a Figure object with the candlestick data and layout
fig = go.Figure(data=candle_data, layout=layout)

# Display the figure
fig.show()
```

# 5.. Analysing closing price (on normal scale & log-scale)

```python
#lets first set "date" as your index to get date on "x-indexes"
data.set_index('Date' , inplace=True)
```

```python
data
```

```python
# Create a Scatter trace with the index values as x and the 'Close' column as y
trace = go.Scatter(x=data.index, y=data['Close'])

# Create a Figure object with the Scatter trace
fig = go.Figure(data=trace)

# Customize the layout of the figure
fig.update_layout(
    title='Close Prices',  # Set the title of the figure
    xaxis_title='Date',    # Set the title of the x-axis
    yaxis_title='Price'    # Set the title of the y-axis
)

# Display the figure
fig.show()
```

```python
# Add a Scatter trace for the original data without scaling
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], name='No scaling'))

# Add a Scatter trace for the data with logarithmic scaling
fig.add_trace(go.Scatter(x=data.index, y=np.log1p(data['Close']), name='Log scaling'))

# Update the layout of the figure
fig.update_layout(
    title='Comparison of Price Scaling',  # Set the title of the figure
    xaxis_title='Date',                   # Set the title of the x-axis
    yaxis_title='Price',                  # Set the title of the y-axis
    yaxis_type='log'                      # Set the y-axis to logarithmic scale
)

# Display the figure
fig.show()
```

##### Logarithmic price scales are better than linear price scales at showing less severe price increases or decreases..

##### log scale is always better if u have outliers or some high values in data

Some features of the plot above:

    There is an up-ward trend from 2016 for each graph
    There is no seasonality
    There are no outliers

# 6.. Analyse closing Price on Yearly , Quarterly , monthly basis !

```python
import plotly.graph_objects as go

fig = go.Figure()

# Yearly resampled mean
fig.add_trace(go.Scatter(x=data.resample('Y')['Close'].mean().index, y=data.resample('Y')['Close'].mean(),
                         name='Yearly', mode='lines+markers', line=dict(width=2), marker=dict(size=6)))

# Quarterly resampled mean
fig.add_trace(go.Scatter(x=data.resample('Q')['Close'].mean().index, y=data.resample('Q')['Close'].mean(),
                         name='Quarterly', mode='lines+markers', line=dict(width=2), marker=dict(size=6)))

# Monthly resampled mean
fig.add_trace(go.Scatter(x=data.resample('M')['Close'].mean().index, y=data.resample('M')['Close'].mean(),
                         name='Monthly', mode='lines+markers', line=dict(width=2), marker=dict(size=6)))

# Update line width and marker size for all traces
fig.update_traces(
    line=dict(width=3),       # Increase line width to 3
    marker=dict(size=8)       # Increase marker size to 8
)

fig.update_layout(
    title='Resampled Mean of Close Prices',   # Set the title of the figure
    xaxis_title='Date',                       # Set the title of the x-axis
    yaxis_title='Mean Close Price',            # Set the title of the y-axis
    hovermode="x unified"                      # Enable x-axis hover data for all traces
)

fig.show()
```
